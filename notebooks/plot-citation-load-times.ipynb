```python
import pandas as pd

# Load the first CSV file
df1 = pd.read_csv('timing-citations-with-keys.csv')

# Load the second CSV file
df2 = pd.read_csv('timing-citations-without-keys.csv')
```

```python
df1['TimeDiff1'] = pd.to_timedelta(df1['TimeDiff1'])
df2['TimeDiff2'] = pd.to_timedelta(df2['TimeDiff2'])
```

```python
# Merge the two dataframes on the 'Counts' column
combined_df = pd.merge(df1, df2, on='Counts')
```

```python
combined_df['TimeDiff1'] = combined_df['TimeDiff1'].dt.total_seconds()
combined_df['TimeDiff2'] = combined_df['TimeDiff2'].dt.total_seconds()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(combined_df['Counts'], combined_df['TimeDiff1'], label='TimeDiff1')
plt.plot(combined_df['Counts'], combined_df['TimeDiff2'], label='TimeDiff2')
plt.xlabel('Counts')
plt.ylabel('Time Differences (seconds)')
plt.title('Time Differences vs. Counts')
plt.legend()
plt.show()
```
